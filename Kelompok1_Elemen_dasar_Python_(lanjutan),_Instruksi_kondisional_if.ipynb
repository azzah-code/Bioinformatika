# Dasar If
* If adalah perintah percabangan pada Python.
* If digunakan untuk menjalankan suatu aksi hanya jika kondisi bernilai benar atau True.
* Program akan mengecek kondisi terlebih dahulu.
* Jika kondisi benar maka program menjalankan perintah di dalam if.
* Jika kondisi salah maka perintah tersebut dilewati.
* Percabangan membantu program membuat keputusan.
* Contoh keputusan dalam program.
1. Menentukan kelulusan.
2. Menentukan diskon.
3. Menentukan apakah seseorang boleh melakukan sesuatu.
4. Dalam bioinformatika, if juga digunakan untuk membantu analisis data biologis.
* Contoh penggunaan dalam bioinformatika.
1. Mengecek apakah sekuens DNA memiliki start codon ATG.
2. Mengecek apakah panjang sekuens DNA cukup untuk dianalisis.
3. Mengecek apakah nilai GC content suatu DNA tinggi atau rendah.
4. Mengecek apakah kualitas data sequencing cukup baik untuk diproses lebih lanjut.



```python
dna_seq = "ACGTATGGGCTA"

if "ATG" in dna_seq:
    print("Start codon ditemukan. Analisis gen dapat dimulai.")
```

# Struktur Dasar If dalam Python
* If memiliki struktur sederhana.
* Program mengecek kondisi terlebih dahulu.
* Jika kondisi benar maka program menjalankan aksi.
* Kondisi adalah pernyataan logika.
* Kondisi menghasilkan nilai True atau False.
* Aksi adalah perintah yang dijalankan jika kondisi True.
* Struktur dasar

if kondisi:

aksi

```python
dna_seq = "ATGCGTACGTTAGC"

if len(dna_seq) > 10:
    print("Panjang sekuens DNA cukup untuk dianalisis.")
```

# Cara Kerja Percabangan
1. Program membaca nilai variabel
2. Program mengevaluasi kondisi
3. Jika True maka perintah dijalankan
4. Jika False maka perintah dilewati

```python
dna_seq = "ATGCGCGTAC"

gc_count = dna_seq.count("G") + dna_seq.count("C")
gc_content = (gc_count / len(dna_seq)) * 100

if gc_content > 50:
    print("GC content tinggi")
```

# If-Else & If-Elif-Else
**If-Else Statement**


*   if-else digunakan untuk menentukan dua kemungkinan kondisi.
*   Jika kondisi bernilai benar, program menjalankan bagian if.
*   Jika kondisi bernilai salah, program menjalankan bagian else.






```python
if kondisi:
    aksi_jika_benar
else:
    aksi_jika_salah
```

```python
suhu = 37

if suhu == 37:
    print("Suhu optimal untuk enzim manusia")
else:
    print("Suhu tidak optimal")
```

**If-Elif-Else Statement**

*   if-elif-else digunakan ketika terdapat lebih dari dua kondisi yang perlu diperiksa.
*   Program akan mengecek kondisi dari atas ke bawah.
*   Ketika satu kondisi terpenuhi, kondisi berikutnya tidak akan diperiksa.

```python
if kondisi1:
    aksi1
elif kondisi2:
    aksi2
else:
    aksi_lainnya
```

```python
ph = 6

if ph < 7:
    print("Larutan bersifat asam")
elif ph == 7:
    print("Larutan netral")
else:
    print("Larutan bersifat basa")
```

# Nested If & Operator Logika

**NESTED IF**



*  **Nested if** adalah kondisi if yang berada **di dalam if lain.**
*   Digunakan untuk **pengecekan bertahap / berlapis.**

```python
umur = 19
punya_ktp = True

if umur >= 17:
    if punya_ktp:
        print("Boleh memilih")
    else:
        print("Tidak boleh memilih karena tidak punya KTP")
else:
    print("Belum cukup umur")
```

```python
#coba identasi dan pindah baris
umur = 19
punya_ktp = True

if umur >= 17:
 if punya_ktp:
  print("Boleh memilih")
 else:
  print("""Tidak boleh memilih
  karena tidak punya KTP""")
else:
  print("Belum cukup umur")
```

**OPERATOR LOGIKA**

**a. AND (dan)**

Operator AND pada python dapat dilakukan menggunakan function and() atau &.

 Nilai kebenaran operator AND adalah **benar ketika kedua operand bernilai benar.**

```python
umur = 20
punya_ktp = True

if umur >= 17 and punya_ktp:
    print("Boleh memilih")
```

**b. OR (atau)**

Operator OR pada python dapat dilakukan menggunakan fungsi or() atau |.

Operator OR mempunyai **nilai kebenaran salah saat kedua operand bernilai salah.**



```python
nilai = 70
remedial = True

if nilai >= 75 or remedial:
    print("Lulus")
```

c. **NOT (tidak)**

Membalik kondisi

```python
punya_ktp = False

if not punya_ktp:
    print("Tidak punya KTP")
```
